```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Funcin para aplicar filtro gaussiano
def aplicar_filtro_gaussiano(imagen, kernel_size=5):
    return cv2.GaussianBlur(imagen, (kernel_size, kernel_size), 0)

# Funcin para modificar brillo y contraste
def ajustar_brillo_contraste(imagen, alpha=1.0, beta=0):
    return cv2.convertScaleAbs(imagen, alpha=alpha, beta=beta)

# Funcin para aplicar ecualizacin de histograma
def ecualizar_histograma(imagen):
    return cv2.equalizeHist(imagen)

# Funcin para calcular y mostrar histogramas
def mostrar_histograma(imagen, titulo):
    hist = cv2.calcHist([imagen], [0], None, [256], [0, 256])  # Histograma de la imagen en escala de grises
    plt.plot(hist, color='black')
    plt.xlim([0, 256])
    plt.title(titulo)
    plt.xlabel("Intensidad")
    plt.ylabel("Frecuencia")

# Rutas de las imgenes
rutas_imagenes = [
    r"C:\Users\hgera\Downloads\DetectarComponentes\Resistencia.png",
    r"C:\Users\hgera\Downloads\DetectarComponentes\Resistencia2.png",
    r"C:\Users\hgera\Downloads\DetectarComponentes\Resistencia3.png",
    r"C:\Users\hgera\Downloads\DetectarComponentes\Resistencia4.png"
]

# Parmetros de ajuste
alpha = 2      # Contraste para imagen original
beta = -500      # Brillo para imagen original
alpha_eq = 2 # Contraste para imagen ecualizada
beta_eq = -255  # Brillo para imagen ecualizada
kernel_size = 11  # Tamao del kernel para el filtro gaussiano

# Crear la figura para mostrar imgenes
plt.figure(figsize=(15, 25))

# Procesar cada imagen y mostrarla en columnas
for i, ruta in enumerate(rutas_imagenes):
    # Cargar imagen en escala de grises
    imagen_gris = cv2.imread(ruta, cv2.IMREAD_GRAYSCALE)
    
    if imagen_gris is None:
        print(f"Error: No se pudo cargar la imagen en {ruta}")
        continue  # Saltar esta imagen y continuar con las dems

    # Aplicar filtro gaussiano
    imagen_suavizada = aplicar_filtro_gaussiano(imagen_gris, kernel_size=kernel_size)

    # Aplicar brillo y contraste
    imagen_brillo_contraste = ajustar_brillo_contraste(imagen_suavizada, alpha=alpha, beta=beta)
    
    # Aplicar ecualizacin de histograma
    imagen_ecualizada = ecualizar_histograma(imagen_suavizada)
    
    # Aplicar brillo y contraste sobre la imagen ecualizada
    imagen_eq_brillo_contraste = ajustar_brillo_contraste(imagen_ecualizada, alpha=alpha_eq, beta=beta_eq)

    # Mostrar imgenes y histogramas
    plt.subplot(len(rutas_imagenes), 6, i * 6 + 1)
    plt.imshow(imagen_suavizada, cmap='gray')
    plt.title(f"Filtro Gaussiano - {i+1}")
    plt.axis("off")

    plt.subplot(len(rutas_imagenes), 6, i * 6 + 2)
    plt.imshow(imagen_brillo_contraste, cmap='gray')
    plt.title(f"Brillo/Contraste - {i+1}")
    plt.axis("off")

    plt.subplot(len(rutas_imagenes), 6, i * 6 + 4)
    plt.imshow(imagen_eq_brillo_contraste, cmap='gray')
    plt.title(f"Eq. Hist + B/C - {i+1}")
    plt.axis("off")

    # Mostrar histogramas de brillo/contraste
    plt.subplot(len(rutas_imagenes), 6, i * 6 + 3)
    mostrar_histograma(imagen_brillo_contraste, f"Histograma B/C - {i+1}")

    # Mostrar histogramas de ecualizacin + brillo/contraste
    plt.subplot(len(rutas_imagenes), 6, i * 6 + 5)
    mostrar_histograma(imagen_eq_brillo_contraste, f"Histograma Eq. + B/C - {i+1}")

plt.tight_layout()
plt.show()
```
